### Splitting Data

```python
sections = [
    "Identification",
    "Hazards identification",
    "Composition/information on ingredients",
    "First aid measures",
    "Firefighting measures",
    "Accidental release measures",
    "Handling and storage",
    "Exposure controls/personal protection",
    "Physical and chemical properties",
    "Stability and reactivity",
    "Toxicological information",
    "Ecological information",
    "Disposal considerations",
    "Transport information",
    "Regulatory information",
    "Other information"
]
```

```python
def split_sections(text, sections):
    split_data = {}
    product_name, supplier_name = None, None
    
    for i in range(len(sections)):
        start = re.search(sections[i], text, re.IGNORECASE)
        end = re.search(sections[i + 1], text, re.IGNORECASE) if i + 1 < len(sections) else None
        
        if start:
            start_pos = start.start()
            end_pos = end.start() if end else len(text)
            section_text = text[start_pos:end_pos].strip()
            
            # Extract product and supplier name from the "Identification" section
            if sections[i].lower() == "identification":
                product_match = re.search(r'Product name\s*:\s*(.*)', section_text, re.IGNORECASE)
                supplier_match = re.search(r'Company name of supplier\s*(.*)', section_text, re.IGNORECASE)
                
                product_name = product_match.group(1).strip() if product_match else ""
                supplier_name = supplier_match.group(1).strip() if supplier_match else ""
            
            split_data[sections[i]] = section_text
    
    return split_data, product_name, supplier_name
```

```python
sample_split_df = pd.DataFrame(columns=['Product Name', 'Supplier Name'] + sections)
```

```python
for line in content:
    doc_sec, product_name, supplier_name = split_sections(line, sections)
    doc_sec['Product Name'] = product_name
    doc_sec['Supplier Name'] = supplier_name
    doc_sec_df = pd.DataFrame([doc_sec])
    sample_split_df = pd.concat([sample_split_df, doc_sec_df], axis=0, ignore_index=True)
```

# Embedding and ChromaDB

```python
import os
from langchain.embeddings import OpenAIEmbeddings
import chromadb
import getpass
from langchain.vectorstores import Chroma

# Set your OpenAI API key directly in the code
os.environ["OPENAI_API_KEY"] = getpass.getpass("Enter your OpenAI API key: ")
```

```python
import os
import chromadb
from langchain.embeddings import HuggingFaceEmbeddings
import pandas as pd

# Load the Excel file where split sections are stored
split_df = pd.read_excel("Sample_Split_Data_Sec.xlsx")  # Ensure the path is correct

# Define the columns of interest based on the structure of your data
columns_of_interest = [
    'Identification',
    'Hazards identification',
    'Composition/information on ingredients',
    'First aid measures',
    'Firefighting measures',
    'Accidental release measures',
    'Handling and storage',
    'Exposure controls/personal protection',
    'Physical and chemical properties',
    'Stability and reactivity',
    'Toxicological information',
    'Ecological information',
    'Disposal considerations',
    'Transport information',
    'Regulatory information',
    'Other information'
]

# Initialize ChromaDB client
client = chromadb.Client()

# Try to get the collection or create it if it does not exist
try:
    collection = client.create_collection(name="chemical_data_embeddings_v3")
except Exception:
    collection = client.get_collection(name="chemical_data_embeddings_v3")

# Initialize Hugging Face Embeddings using the model name
hf_embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Loop through each row in the DataFrame and add the data
for index, row in split_df.iterrows():
    # Combine text from all the sections listed in columns_of_interest
    document_content = " ".join([str(row.get(col, '')) for col in columns_of_interest])
    
    # Generate embedding for the combined document content
    embedding = hf_embeddings.embed_query(document_content)
    
    # Convert the list of sections to a comma-separated string
    sections_present_str = ", ".join([col for col in columns_of_interest if pd.notna(row[col])])
    
    # Store the document and its metadata in ChromaDB
    collection.add(
        documents=[document_content],
        embeddings=[embedding],
        metadatas=[{
            "product_name": row['Product Name'],
            "supplier_name": row['Supplier Name'],
            "file_name": f"Document_{index}",  # Placeholder for document identification
            "sections_present": sections_present_str  # Convert list to a comma-separated string
        }],
        ids=[str(index)]
    )

print("All columns processed, and embeddings stored in ChromaDB.")
```

# Storing the chromaDB

```python
import json

# Retrieve all the data from the ChromaDB collection
all_documents = collection.get()["documents"]
all_embeddings = collection.get()["embeddings"]
all_metadatas = collection.get()["metadatas"]
all_ids = collection.get()["ids"]

# Combine everything into a dictionary
chroma_data = {
    "documents": all_documents,
    "embeddings": all_embeddings,
    "metadatas": all_metadatas,
    "ids": all_ids
}

# Write the dictionary to a JSON file
with open("chroma_db_export.json", "w") as f:
    json.dump(chroma_data, f)

print("ChromaDB collection exported to chroma_db_export.json.")
```

# Setting up the retrieval process

```python
import os
import chromadb
from langchain.vectorstores import Chroma
from langchain_openai import ChatOpenAI  # Updated import for ChatOpenAI
from langchain.chains import RetrievalQA
from langchain.embeddings import HuggingFaceEmbeddings
import getpass
```

```python
os.environ["OPENAI_API_KEY"] = getpass.getpass("Enter your OpenAI API key: ")   
```

```python
# Initialize ChromaDB client
client = chromadb.Client()

# Access the existing collection
collection_name = "chemical_data_embeddings_v3"
collection = client.get_collection(name=collection_name)

# Initialize the embedding function using Hugging Face
hf_embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
# Set up the vector store using the ChromaDB collection and the embedding function
vector_store = Chroma(
    embedding_function=hf_embeddings,
    client=client,
    collection_name=collection_name
)

# Set up the chat model
llm = ChatOpenAI(model="gpt-4o")
```

```python
# Set up the RetrievalQA chain

retrieval_qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=vector_store.as_retriever()
)
```

```python
# Define a query to retrieve information
query = "Please provide a list of chemicals containing Category 1 GHS Classifications"
```

```python

# Execute the query using the QA chain
result = retrieval_qa_chain({"query": query})

# Display the result
print(result['result'])
```

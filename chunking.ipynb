### Splitting Data

```python
sections = [
    "Identification",
    "Hazards identification",
    "Composition/information on ingredients",
    "First aid measures",
    "Firefighting measures",
    "Accidental release measures",
    "Handling and storage",
    "Exposure controls/personal protection",
    "Physical and chemical properties",
    "Stability and reactivity",
    "Toxicological information",
    "Ecological information",
    "Disposal considerations",
    "Transport information",
    "Regulatory information",
    "Other information"
]
```

```python
def split_sections(text, sections):
    split_data = {}
    product_name, supplier_name = None, None
    
    for i in range(len(sections)):
        start = re.search(sections[i], text, re.IGNORECASE)
        end = re.search(sections[i + 1], text, re.IGNORECASE) if i + 1 < len(sections) else None
        
        if start:
            start_pos = start.start()
            end_pos = end.start() if end else len(text)
            section_text = text[start_pos:end_pos].strip()
            
            # Extract product and supplier name from the "Identification" section
            if sections[i].lower() == "identification":
                product_match = re.search(r'Product name\s*:\s*(.*)', section_text, re.IGNORECASE)
                supplier_match = re.search(r'Company name of supplier\s*(.*)', section_text, re.IGNORECASE)
                
                product_name = product_match.group(1).strip() if product_match else ""
                supplier_name = supplier_match.group(1).strip() if supplier_match else ""
            
            split_data[sections[i]] = section_text
    
    return split_data, product_name, supplier_name
```

# Fuzzy Logic

## **Trial method used for supplier name extraction. Please run the above code for overll better results**

```python
import re
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
supplier_patterns = [
    "Company name of supplier",
    "Supplier:",
    "Company:",
    "Manufacturer:",
    "Produced by:",
    "Manufacturer or supplier's details Company name of supplier",
    "Name of the supplier:",
    "Company identification: Manufacturer's Name",
    "S u p p l i e r",
    "Details of supplier of safety data sheet: Name of manufacturer/supplier:",
    "Manufacturer/Importer/Supplier/Distributor information Company Name:"
]
```

```python
def find_section_start(text, section_titles):
    for title in section_titles:
        match = re.search(re.escape(title), text, re.IGNORECASE)
        if match:
            return match.start()
    return None
```

```python
def extract_supplier_name_fuzzy(text):
    SIMILARITY_THRESHOLD = 30
    best_match = None
    best_score = 0
    lines = text.splitlines()
    
    for pattern in supplier_patterns:
        match = process.extractOne(pattern, lines, scorer=fuzz.ratio)
        
        if match and match[1] >= SIMILARITY_THRESHOLD and match[1] > best_score:
            best_match, best_score = match
    
    if best_match:
        matched_line = best_match
        
        supplier_name_pattern = re.escape(matched_line) + r'\s*[:\-]?\s*(.*)'
        supplier_match = re.search(supplier_name_pattern, text, re.IGNORECASE)
        
        if supplier_match:
            supplier_name = supplier_match.group(1).strip()
            
            if supplier_name and supplier_name.lower() not in ["", "n/a", "not applicable"]:
                return supplier_name
    
    return ""
```

```python
def split_sections(text, section_mapping):
    split_data = {}
    product_name, supplier_name = None, None
    
    sections = list(section_mapping.keys())
    for i in range(len(sections)):
        current_section = sections[i]
        next_section = sections[i + 1] if i + 1 < len(sections) else None
        
        start = find_section_start(text, section_mapping[current_section])
        end = find_section_start(text, section_mapping[next_section]) if next_section else None
        
        if start is not None:
            section_text = text[start:end].strip() if end else text[start:].strip()
            
            if current_section == 1:
                product_match = re.search(r'Product name\s*:\s*(.*)', section_text, re.IGNORECASE)
                product_name = product_match.group(1).strip() if product_match else ""
                
                supplier_name = extract_supplier_name_fuzzy(section_text)
            
            split_data[current_section] = section_text
    
    return split_data, product_name, supplier_name
```
